```python
# important modules
import numpy as np
import cv2
import matplotlib.pyplot as plt
from matplotlib import cm
from Module import Module as md
```

```python

#采用直方图二值化
hist = cv2.equalizeHist(spectrum)
# hist = cv2.equalizeHist(binary)

plt.subplot(121)
plt.hist(spectrum.ravel(), 256)
plt.subplot(122)
plt.hist(hist.ravel(), 256)
plt.show()
```

```python
mdata=np.hstack((binary,hist))
plt.imshow(mdata,cmap=cm.gray)
a=cv2.imwrite("./out/row_ana.png",mdata)
```
